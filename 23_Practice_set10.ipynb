# 1. Create a class “Programmer” for storing information of few programmers working at Microsoft.

# 2. Write a class “Calculator” capable of finding square, cube and square root of a number. 


```python
import math

class calculator:
    def __init__(self, n):
        self.n = n

    def square(self):
        print(f"The Square of {self.n} is : {self.n * self.n}")

    def cube(self):
        print(f"The Cube of {self.n} is : {self.n * self.n * self.n}")
    
    def root(self):
        print(f"The Square root of {self.n} is : {math.sqrt(self.n)}")   #using in-built module
        print(f"The Square root of {self.n} is : {self.n**1/2}") #without using in-build module

a = calculator(4)
a.square()
a.cube()
a.root()
```

# 4. Add a static method in problem 2, to greet the user with hello. 


```python
import math

class calculator:
    def __init__(self, n):
        self.n = n
    
    def square(self):
        print(f"The Square of {self.n} is : {self.n * self.n}")

    def cube(self):
        print(f"The Cube of {self.n} is : {self.n * self.n * self.n}")
    
    def root(self):
        print(f"The Square root of {self.n} is : {math.sqrt(self.n)}")   #using in-built module
        print(f"The Square root of {self.n} is : {self.n**1/2}") #without using in-build module

    @staticmethod
    def greet():
        print("Hello User")

a = calculator(4)
a.square()
a.cube()
a.root()
calculator.greet()

```

# 5. Write a Class ‘Train’ which has methods to book a ticket, get status (no of seats) and get fare information of train running under Indian Railways. 


```python
from random import randint 

class trian:
    
    def __init__(self,name, tNo, fro, to):
        self.tNo = tNo
        self.fro = fro
        self.to = to
        self.name = name
        

    def book(self):
        print(f"Hello Sir/Madam : {self.name}.")
        print(f"Your Train {self.tNo} from {self.fro} to {self.to} is Booked Successfully.")
        print("Happy Journey :)")
        print("")

    def getStatus(self):
        print(f"Your Train {self.tNo} from {self.fro} to {self.to} is on Time...")
        print("")

    def fare(self):
        print(f"Hello Sir/Madam : {self.name}.")
        print(f"Your Train {self.tNo} from {self.fro} to {self.to}.")
        print(f"Ticket Number : {randint(101,999)}")
        print(f"Seat Number : {randint(1,300)}")
        print("")

t= trian("Harsh Patil", 110212, "Mumbai", "Pune")
t.book()
t.fare()
t.getStatus()

t2 = trian("Anushka", 221011, "Pune", "Goa")
t2.fare()
t2.getStatus()
```
