# LAMBDA FUNCTIONS
Function created using an expression using ‘lambda’ keyword.


```python
#syntax
lambda arguments:expressions
```

```python
a = int(input("Enter a number : "))
sqaure = lambda a : a*a
print(sqaure(a))
```

```python
sum  = lambda a,b,c:a+b+c
sum(1,2,3)
```

# JOIN METHOD (STRINGS)
Creates a string from iterable objects

```python
l = ["apple", "mango", "banana"]
result = " and ".join(l)
print(result)
```

# MAP, FILTER & REDUCE
Map applies a function to all the items in an input_list.

```python
l = [1,2,3,4,5,6]

square = lambda x : x*x

sqlist = map(sqaure, l)
print(list(sqlist))
```

Filter

Filter creates a list of items for which the function returns true.

```python
l = [1,2,3,4,5,6]

def even(n):
    if (n%2 == 0):
        return True
    return False

onlyEven = filter(even, l)
print(list(onlyEven))
```

# Reduce

Reduce applies a rolling computation to sequential pair of elements.

```python
from functools import reduce

l = [1,2,3,4,5,6]

def sum(a,b):
    return a+b

print(reduce(sum, l))
```
