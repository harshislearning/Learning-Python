## Python Inheritance

Inheritance allows us to define a class that inherits all the methods and properties from another class.

Parent class is the class being inherited from, also called base class.

Child class is the class that inherits from another class, also called derived class.

```python
class Employee:  #base class
    pass

class Programmer(Employee):         # Derived clas or child class
    pass
```

We can use the method and attributes of ‘Employee’ in ‘Programmer’ object. 

Also, we can overwrite or add new attributes and methods in ‘Programmer’ class. 





**TYPES OF INHERITANCE** 

• Single inheritance 

• Multiple inheritance 

• Multilevel inheritance 

# SINGLE INHERITANCE 
Single inheritance occurs when child class inherits only a single parent class. 

```python
class Employee:  #base class or parent class
    pass

class Programmer(Employee):         # Derived clas or child class
    pass
```

#   MULTIPLE INHERITANCE 
Multiple Inheritance occurs when the child class inherits from more than one parent classes



```python
class Employee:  #base class or parent class
    pass

class coder:       #2nd base class or parent class
    pass

class Programmer(Employee, coder):         # Derived class or child class
    pass
```

# MULTILEVEL INHERITANCE 
When a child class becomes a parent for another child class.

Parent-->child1-->child2

```python
class Employee:  #base class or parent class
    def __init__(self):
        print("Constructor of Employee")
    a = 1
    

class coder(Employee):       #Derived class or child class
    def __init__(self):
        print("Constructor of coder")
    b = 2
    

class Programmer(coder):         #2nd Derived class or child class
    def __init__(self):
        print("Constructor of Programmer")
    c = 3
    

# o1 = Employee()
# print(o1.a)
# # print(o1.b)   ---error
# # print(o1.c)   ---error

# o2 = coder()
# print(o2.a)
# print(o2.b)   
# # print(o1.c)   ----error

o3 = Programmer()
print(o3.a)    
print(o3.b)
print(o3.c)
```

#   SUPER() METHOD  
super() method is used to access the methods of a super class in the derived class. 

```python
class Employee:  
    a = 1
    def __init__(self):
        print(f"Constructor of Employee : {self.a}")
    
    

class coder(Employee):     
    b = 2  
    def __init__(self):
        super().__init__()         # super method
        print(f"Constructor of coder : {self.a} {self.b}")
    
    

class Programmer(coder):  
    c = 3       
    def __init__(self):
        super().__init__()           #super method
        print(f"Constructor of Programmer : {self.a} {self.b} {self.c}")
    
    

o3 = Programmer()
```

# Class method
A class method is a method which is bound to the class and not the object of the class. 
@classmethod decorator is used to create a class method. 

```python
class example:
    a = 1
    def show(self):
        print(f"The Class attribute :{self.a}")


object = example()
object.a = 99
object.show()
```

**I dont want to change my class attibute**

```python
class example:
    a = 1

    @classmethod
    def cl(cls):
        print(f"The Class attribute : {cls.a}")


    def obj(self):
        print(f"The Object attribute : {self.a}")


object = example()
object.a = 99
object.cl()
object.obj()
```

**Hence when we use @classmethod the class attribute wont change.**
