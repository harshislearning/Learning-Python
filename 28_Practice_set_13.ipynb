# 3. A list contains the multiplication table of 7. write a program to convert it to vertical string of same numbers.
7 14 .....

```python
seven = [str(7*i) for i in range(1,11)]

s = "\n".join(seven)
print(s)
```

# 4. Write a program to filter a list of numbers which are divisible by 5.

```python
a = [233,343,45,67,23,456,788,653,34,50,70,35,234,56,55,65,12,15,45,66,88,65]

def dev5(n):
    if (n%5 == 0):
        return True
    return False

f = list(filter(dev5, a))
print(list(f))
```

# 5. Write a program to find the maximum of the numbers in a list using the reduce function.


```python
from functools import reduce

a = [233,343,45,67,23,456,788,653,34,50,70,35,234,56,55,999,65,12,15,45,66,88,65]

def greater(a,b):
    if(a > b):
        return a
    return b

print(f"Greater Number from list is : {reduce(greater, a)}")

```
