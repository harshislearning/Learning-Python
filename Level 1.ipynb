**Question 4:** Write a Python program to find the largest of three numbers.

```python
i1 = int(input("Enter Number: "))
i2 = int(input("Enter Number: "))
i3 = int(input("Enter Number: "))

if (i1 > i2) and (i1 > i3):
    print(f"{i1} is largest number.")
elif (i2 > i1) and (i2 > i3):
    print(f"{i2} is largest number.")
elif (i3 > i2) and (i3 > i1):
    print(f"{i3} is largest number.")
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
num = int(input("Enter Number: "))

n = 1
for i in range(1,num+1):
    n = n*i

print(f"Factorial of {num} is {n}.")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
year = int(input("Enter number :"))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} It is a leap year")
else:
    print(f"{year} It is not a leap year")
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
s = input("Enter a string: ")

if s == s[::-1]:
    print(f"Given string : '{s}' is a palindrome")
else:
    print(f"Given string : '{s}'  is not a palindrome")
```
