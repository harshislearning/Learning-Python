
### Assignment 3: Set Operations

Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.


```python
set1 = {i for i in range(1,6)}
set2 = {i for i in range(1,11) if i %2 ==0}

print(f"1st List : {set1}")
print(f"2nd List : {set2}")

u = set1.union(set2)
print(f"union : {u} ")

comon = set1.intersection(set2)
print(f"Intersection : {comon} ")

diff = set1.difference(set2)
print(f"Difference : {diff} ")

all_diff = set1.symmetric_difference(set2)
print(f"Symmetric Diff : {all_diff} ")
```
