 ## FILE I/O

 The random-access memory(ram) is volatile, and all its contents are lost once a program terminates. In order to persist(save) the data forever, we use files. 

 A file is data stored in a storage device. A python program can talk to the file by reading 
content from it and writing content to it.

TYPE OF FILES. 
There are 2 types of files: 

1. Text files (.txt, .c, etc) 
2. Binary files (.jpg, .dat, etc) 

Python has a lot of functions for reading, updating, and deleting files.

| Mode   | Meaning           |
| ------ | ----------------- |
| `"r"`  | Read (default)    |
| `"w"`  | Write (overwrite) |
| `"a"`  | Append            |
| `"x"`  | Create new file   |
| `"rb"` | Read binary       |
| `"wb"` | Write binary      |


**Opening a file and closing it:**

```python
f = open("20_text.txt", "r") #Opening a file, "r" is a read mode by default its on read mode only  
data = f.read()         #reading the file 
print(data)             
f.close()               #closing
```

**Using "w"(write) mode to overwrite the text.**

"w" mode creates the file if it doesnâ€™t exist

"w" overwrites existing content

```python
st = "I want to add this text in text file"

f = open("20_text.txt", "w") 
f.write(st)                       
f.close()               

f = open("20_text.txt", "r")   
data = f.read()     
print(data)    
f.close()  

```

**Using "a" append, to add data without overwriting**

```python
new_st ="""\nHello
This is example 
Hope you have a great day."""

f = open("20_text.txt", "a")
f.write(new_st)  
f.close()  

f = open("20_text.txt", "r")   
data = f.read()     
print(data)    
f.close()  
```

**Using "with" statement**

No need to manually close the file.

Python automatically closes the file.


```python
with open("20_text.txt", "r") as f:
    data = f.read()
    
print(data)
```
